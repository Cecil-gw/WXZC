```python
power = 0
intelligence = 0
agile = 0
total_point = 10
MAX_SUM = 20 

print("========属性加点========")
print(f"初始属性：力量{power},智力{intelligence},敏捷{agile},可分配点数{total_point}")
print("单次加点范围限制：-10 ~ 20")

while True:
    # 总属性上限判定
    if power + intelligence + agile >= MAX_SUM:
        print(f"\n总属性总和达到上限{MAX_SUM}，自动结束加点！")
        break

    # 输入选择属性
    choice = int(input("\n请输入要加点的属性编号(0退出)："))

    if choice == 0:
        break

    # 分配力量点数
    if choice == 1:
        while True:
            add = int(input(f"输入要增加的力量点数(-10~{total_point})："))
            # 拆分判断：负数只限制最低-10；正数不能超过剩余点数
            if add < 0:
                if -10 <= add <= 0:
                    break
                else:
                    print("扣除属性单次最多-10！")
            else:
                if 0 <= add <= total_point:
                    break
                else:
                    print(f"加点不能超过当前剩余点数{total_point}！")
        
        power += add
        total_point -= add
        print(f"力量{add:+d}，当前力量：{power}，剩余点数：{total_point}")

    # 分配智力点数
    elif choice == 2:
        while True:
            add = int(input(f"输入要增加的智力点数(-10~{total_point})："))
            if add < 0:
                if -10 <= add <= 0:
                    break
                else:
                    print("扣除属性单次最多-10！")
            else:
                if 0 <= add <= total_point:
                    break
                else:
                    print(f"加点不能超过当前剩余点数{total_point}！")
            
        intelligence += add
        total_point -= add
        print(f"智力{add:+d}，当前智力：{intelligence}，剩余点数：{total_point}")

    # 分配敏捷点数
    elif choice == 3:
        while True:
            add = int(input(f"输入要增加的敏捷点数(-10~{total_point})："))
            if add < 0:
                if -10 <= add <= 0:
                    break
                else:
                    print("扣除属性单次最多-10！")
            else:
                if 0 <= add <= total_point:
                    break
                else:
                    print(f"加点不能超过当前剩余点数{total_point}！")
            
        agile += add
        total_point -= add
        print(f"敏捷{add:+d}，当前敏捷：{agile}，剩余点数：{total_point}")

    else:
        print("输入错误！只能输入0、1、2、3")

# 退出循环后：数值汇总+转换打印
print("\n===== 最终角色属性 =====")
print(f"力量(power) = {power}")
print(f"智力(intelligence) = {intelligence}")
print(f"敏捷(agile) = {agile}")
sum_attr = power + intelligence + agile
print(f"三属性总和：{sum_attr}")

# 定义描述字符串
strength_desc = ""
smart_desc = ""
slight_desc = ""

# 力量 match 判断
match power:
    case x if x >= 15:
        strength_desc = "勇武の"
    case x if x >= 10:
        strength_desc = "善武的"
    case x if x >= 5:
        strength_desc = "强壮的"
    case x if x >= 0:
        strength_desc = "普通身板的"
    case x if x >= -5:
        strength_desc = "瘦弱的"
    case _:
        strength_desc = "羸弱的"

# 智力 同模板
match intelligence:
    case x if x >= 15:
        smart_desc = "博学の"
    case x if x >= 10:
        smart_desc = "聪慧的"
    case x if x >= 5:
        smart_desc = "睿智的"
    case x if x >= 0:
        smart_desc = "普通智慧的"
    case x if x >= -5:
        smart_desc = "愚钝的"
    case _:
        smart_desc = "痴呆的"

# 敏捷 同模板
match agile:
    case x if x >= 15:
        slight_desc = "疾影の"
    case x if x >= 10:
        slight_desc = "灵巧的"
    case x if x >= 5:
        slight_desc = "轻快的"
    case x if x >= 0:
        slight_desc = "普通敏捷的"
    case x if x >= -5:
        slight_desc = "笨拙的"
    case _:
        slight_desc = "迟缓的"

# 获取三属性最大值
max_attr = max(power, intelligence, agile)
job = ""
if max_attr == power:
    flag_rash = False
    if power >= 10 and intelligence <= 0:
        flag_rash = True
    if power >= 15:
        if flag_rash:
            job = "莽撞的无畏战士"
        else:
            job = "无畏战士"
    elif power >= 10 and agile >= 8:
        job += "重骑兵"
    elif flag_rash:
        job = "莽撞的"
elif max_attr == intelligence:
    if intelligence >= 15:
        job = "大法师"
    elif intelligence >= 10 and power >= 8:
        job = "战斗法师"
    else:
        job = "术士"
elif max_attr == agile:
    if agile >= 15:
        job = "游侠"
    elif agile >= 12 and power >= 5 and intelligence >= 2:
        job = "刺客"
    else:
        job = "侠客"
else:
    job = "均衡冒险者"

# 同时输出称号+职业
print(f"\n您的角色是：{strength_desc}{smart_desc}{slight_desc}冒险者，职业：{job}")
```

## 5.1 Python 模块 ，包和第三方库

```python
#一、什么是模块（Module）

# 一个 .py 文件就是一个模块。

# 例如：

# project/
# ├── main.py
# ├── math_utils.py

# math_utils.py

def add(a, b):
    return a + b

# main.py

import math_utils

print(math_utils.add(3, 5))

# 输出：

# 8

# 所以：

# 一个 Python 文件（.py）就是一个模块。


# 二、什么是包（Package）

# 如果模块很多，可以放到一个文件夹里管理，这个文件夹就是包。

# 例如：

# project/
# │
# ├── main.py
# │
# └── tools/
#     ├── __init__.py
#     ├── math_utils.py
#     └── string_utils.py

# 其中：

# tools/

# 就是一个包。

# 导入：

# from tools import math_utils

# print(math_utils.add(1, 2))

# 或者：

# from tools.math_utils import add

# print(add(1, 2))
# __init__.py 是什么？

# 以前 Python 中，一个文件夹要成为包，必须有：

# __init__.py

# 这个文件可以是空的：

# # __init__.py

# 它表示：

# 这个文件夹是一个 Python 包。

# （Python 3.3 以后，即使没有 __init__.py，很多情况下也能作为包使用，但考试一般仍按有 __init__.py 来理解。）

# 三、import 导入方式
# ① 导入整个模块
# import math

# 调用：

# math.sqrt(16)
# ② 导入指定内容
# from math import sqrt

# 调用：

# sqrt(16)
# ③ 起别名
# import math as m

# print(m.sqrt(16))

# 或者：

# from math import sqrt as s

# print(s(16))
# ④ 导入所有（一般不推荐）
# from math import *

# 可以直接：

# print(sqrt(16))

# 但容易造成命名冲突，因此实际开发较少使用。

# 四、标准库

# Python 自带的库叫标准库。

# 例如：

# import math
# import random
# import time
# import os
# import datetime

# 这些安装 Python 后就能直接使用。

# 常见标准库：

# 库	功能
# math	数学计算
# random	随机数
# time	时间
# os	操作系统
# sys	Python 运行环境
# datetime	日期时间
# 五、第三方库

# 不是 Python 自带的，需要安装。

# 例如：

# requests（发送网络请求）
# numpy（科学计算）
# pandas（数据分析）
# matplotlib（绘图）
# flask（Web 开发）

# 安装命令：

# pip install requests

# 使用：

# import requests
# 六、pip

# pip 是 Python 的包管理工具。

# 常见命令：

# 安装：

# pip install 库名

# 例如：

# pip install numpy

# 升级：

# pip install --upgrade numpy

# 卸载：

# pip uninstall numpy

# 查看已安装：

# pip list
# 七、标准库和第三方库区别
# 标准库	第三方库
# Python 自带	需要安装
# 不需要 pip	需要 pip install
# 如 math、os	如 numpy、pandas、requests
```

```python
import json
import random
#1.字典初始化角色属性

player={
    "name": "",
    "hp":100,
    "attack":10,
    "defence":5,
    "rate":0.1,
    "level":1
}

enermy_01={
    "name": "sunshine",
    "hp":50,
    "attack":12,
    "defence":3,
    "rate":0.2,
    "level":1
}

#2.装饰器：双倍伤害的挂件

def double_damage(fn):

    def inner(*args, **kwargs):

        normal_damage = fn(*args, **kwargs)

        print("普通伤害：", normal_damage)

        double = normal_damage * 2

        print("🔥触发双倍伤害BUFF！")
        print("最终伤害：", double)

        return double

    return inner

    

#3.lambda建议基础伤害计算

#atk 攻击力 ， rate技能倍率,dfn 防御
damage_cal = lambda atk,rate,dfn : atk*(1+rate)-dfn

#4.生成器：批量产出buff效果，每次轮询一次

def buff_generator ():
    yield "攻击提升"
    yield "暴击率提升"
    yield "格挡提升"

#5.列表推导式，生成合法加点范围

points = [i for i in range (random.randint(1,6))]

#6.通用属性加点函数


def add_point(player,value):
    while value>0:
        print(f"你可以分配的点数有{value}")
        choice=input("你想添加什么属性？hp，atk，dfn or rate?")
        
        number=int(input(f"想分配多少点数？要小于{value}哦"))
        while number>value:
            number=int(input(f"点数不够哦？要小于{value}哦"))
        
        if choice == "hp":
            player["hp"]+=number*50
        elif choice == "atk":
            player["attack"]+=number
        elif choice == "dfn":
            player["defence"]+=number
        elif choice == "rate":
            player["rate"]+=number*0.01
        else:
            break
        value -= number
    return 1

#7.绑定装饰器的技能伤害函数

@double_damage
def skill(atk, rate,dfn):
    # 返回基础伤害
    return damage_cal(atk, rate,dfn)

#8.with实现角色数据存档
def save_player():
    with open("player.json", "w", encoding="utf-8") as file:
        json.dump(player, file, ensure_ascii=False, indent=4)

    print("角色数据保存成功！")

#9.实现对战效果
def fight(player,enermy):
    while(player["hp"]>0 and enermy["hp"] >0):
        # 玩家攻击
        damage = skill(
            player["attack"],
            player["rate"],
            enermy["defence"]
        )

        enermy["hp"] = int(enermy["hp"]-damage)

        print(f"{player['name']}造成{damage}点伤害")
        print(f"{enermy['name']}剩余血量：{enermy['hp']}")

        # 判断敌人是否死亡
        if enermy["hp"] <= 0:
            print("敌人被击败！")
            break

        # 敌人攻击
        damage = damage_cal(
            enermy["attack"],
            enermy["rate"],
            player["defence"]
        )

        player["hp"] = int( player["hp"]-damage)

        print(f"{enermy['name']}造成{damage}点伤害")
        print(f"{player['name']}剩余血量：{player['hp']}")

        # 判断玩家是否死亡
        if player["hp"] <= 0:
            print("你失败了！")
            break

#10.主函数：交互菜单
def main():
    print("======角色信息======")
    print("请输入您的姓名")
    name=input()
    player["name"]=name
    print("name:",player["name"])
    print("hp:",player["hp"])
    print("attack:",player["attack"])
    print("defence:",player["defence"])
    print("rate:",player["rate"])

    print("\n======属性加点======")
    print("可加点：", points[-1])

    add_point(player, points[-1])

    print("\n======BUFF效果======")
    buffs = buff_generator()

    for buff in buffs:
        print("获得BUFF：", buff)

    print("\n======技能伤害======")
    print("敌人属性:")
    print("姓名：", enermy_01["name"])
    print("血量：", enermy_01["hp"])
    print("攻击：", enermy_01["attack"])
    print("防御：", enermy_01["defence"])
    print("暴击率：", enermy_01["rate"])
    print("\n======开始战斗======")
    fight(player, enermy_01)


    print("\n======保存数据======")
    save_player()


#程序入口：
if __name__ == "__main__":
    main()


```
