1.迭代器：列表、字符串出个遍历

```python
def buff_generator():
    buffs = ["力量增幅+5","智力增幅增幅+3","敏捷增幅+2"]
    for buff in buffs:
        yield buff
g = buff_generator()
print("获得buff:", next(g))
print("获得buff:", next(g))
```

## 6.1装饰器
核心作用：不修改原函数内部代码，动态的给函数新增功能，比如会应用日志打印，权限校验，缓存，异常捕获等等.....

```python
def double_effect(func):
    def wrapper(*args, **kwargs):
        print("魔法特效：双倍伤害！")
        res = func(*args, **kwargs)
        return res*2
    return wrapper

@double_effect
def attack_damage(base):
    return base

print("最终伤害：", attack_damage(20))
```

### 6.1.2 日志打印（最常用）

```python
import time  #库名字.函数名(参数or无参数)

def log_record(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        res = func(*args, **kwargs)
        cost = time.time() - start
        print(f"调用{func.__name__}，耗时{cost:.2f}s")
        return res
    return wrapper

@log_record
def calc_damage(base):    
    return base * 1.5

calc_damage(20)
```

### 6.2.3 权限校验/接口限流/放重复提交

```python
def login_required(func):
    def wrapper(user, *args, **kwargs):
        if not user:
            print("请先登录！")
            return
        return func(user,*args, **kwargs)
    return wrapper

@login_required
def delete_item(user, item_id):
    print(f"删除道具{item_id}")

delete_item(True, 124239057)
delete_item(False, 124239057)
```

### 6.2.4 缓存

```python
cache = {(5, 2): 25}
def cache_func(func):
    def wrapper(*args):
        if args in cache:
            print("命中缓存,直接读取")
            return cache[args]
        res = func(*args)
        cache[args] = res
        return res
    return wrapper

@cache_func
def big_calc(a,b):
    print("复杂计算中...")
    return a**b

big_calc(5,2)
print(cache)
```
